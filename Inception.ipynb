```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
inception_v3=tf.keras.applications.InceptionV3(
    weights='imagenet',
    include_top=False

)
```

```python
def load_img(img_path):
  img= tf.io.read_file(img_path)
  img=tf.io.decode_jpeg(img,channels=3)
  img=tf.keras.layers.Resizing(299,299)(img)
  img=img/255.
  return img

def get_feature_vector(img_path):
  img=load_img(img_path)
  img=tf.expand_dims(img,axis=0)
  feature_vector=inception_v3(img)
  return img, feature_vector
```

```python
img, feature_vector=get_feature_vector('/content/lake-192990__340.jpg')

plt.imshow(np.squeeze(img, axis=0))
plt.axis('off')
plt.show()

print()
print("Input size image :", img.shape)
print('Feature vector size :', feature_vector.shape)
```
